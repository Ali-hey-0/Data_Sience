```python
import matplotlib.pyplot as plt  
import pandas as pd    
import numpy as np 


heart_disease = pd.read_csv("../env/csv files/heart-disease.csv")
over_50 = heart_disease[heart_disease["age"] > 50 ]


fig, ax=plt.subplots(figsize=(10,6))
scatterr = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter");  
ax.set(title="Heart Disease And Cholestrol Level" , xlabel="age",ylabel="Cholestrol" ); 

ax.legend(*scatterr.legend_elements() , title="target"); 

ax.axhline(over_50["chol"].mean(),linestyle="--"); 
```

```python
heart_disease = pd.read_csv("../env/csv files/heart-disease.csv")
over_50 = heart_disease[heart_disease["age"] > 50 ]

fig, (ax0, ax1) = plt.subplots(nrows=2,ncols=1,figsize=(10,10)) 

scatteerr = ax0.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"],cmap="winter")
ax0.set(title="Heart Disease And Cholestrol Levels",xlabel="Age",ylabel="Cholestrol") 

ax0.set_xlim([50,80])

ax0.legend(*scatterr.legend_elements() , title="target"); 
ax0.axhline(y=over_50["chol"].mean(),linestyle="--"); 

# customize a1
scatteerr = ax1.scatter(x=over_50["age"],y=over_50["thalach"],c=over_50["target"],cmap="winter")
ax1.set(title="Heart Disease And Max Rate",xlabel="Age",ylabel="Max Heart Rate")
ax1.set_xlim([50,80])
ax1.set_ylim([80,220])

 
ax1.legend(*scatterr.legend_elements() , title="target"); 
ax1.axhline(y=over_50["thalach"].mean(),linestyle="--"); 

fig.suptitle("Heart Disease Analysist",fontsize=16,fontweight="bold"); 
```
